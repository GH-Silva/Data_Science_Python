### Teste do QUI-QUADRADO - Verificar se as duas variáveis são independentes e se as duas variávies não são independentes.

### No teste, o objetivo é configurar ou recusar a hipótese nula.

## Quando o valor de probabilidade H0 for menor que 0.05 (p-value)


```python
df_tabcont = pd.crosstab(df_churn["Churn"], df_churn["Contract"])
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_scores = chi2_contingency(df_tabcont)
```

```python
chi_scores
```

```python
# Número decimal de uma notação cientifica

pd.set_option("display.float_format", lambda: "%.15f"% x)
```

```python
scores_churn = pd.Series(chi_scores[0])
pvalues_churn = pd.Series(chi_scores[1])
```

```python
df_chi_churn = pd.DataFrame({
    "Qui2": scores_churn,
    "PValue": pvalues_churn
})
```

```python
df_chi_churn
```

### Técnica IQR

```python
# 1° quartil e 3° quartil | 3° quartil - 1° quartil

q1_total = df_churn_month["TotalCharges"].quantile(0.25)
q3_total = df_churn_month["TotalCharges"].quantile(0.75)
```

```python
iqr = q3_total - q1_total
iqr
```

```python
lim_inf = q1_total - (iqr*1.5)
lim_sup = q3_total + (iqr* 1.5)
```

```python
lim_inf, lim_sup
```

```python
len(df_churn_month[(df_churn_month["TotalCharges"] < lim_inf) | (df_churn_month["TotalCharges"] > lim_sup)])
```
