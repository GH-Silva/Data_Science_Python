```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
### Criando dados ficticios de preços de ações para diferentes empresas ao longo de trimestres:

empresas = ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D']
trimestres = ['T1', 'T2', 'T3', 'T4']
```

```python
### Valores aleatórios de 0 a 100 para simular os preços das ações:
dados = np.random.rand(4, 4) * 100
```

```python
### Criando um DataFrame com os dados
df = pd.DataFrame(
    dados,
    columns=trimestres,
    index=empresas
)
```

```python
### Criando o Heatmap com a utilização do Seaborn:
plt.Figure(figsize=(8,6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Preços de Ações por Trimestres')
plt.xlabel('Trimestre')
plt.ylabel('Empresas')
plt.show()
```
