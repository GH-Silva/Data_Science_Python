```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Dados ficticios - Pontuações de um teste

pont = np.random.randint(50, 100, 100)
print(pont)
```

```python
### Criando o histograma

plt.Figure(figsize=(8,5))
plt.hist(
    pont,
    bins=10,
    color='skyblue',
    edgecolor='black'
)

### Adicionado rótulos e título ao gráfico

plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.title('Distribuição das Pontuações do Teste')
plt.show()
```
