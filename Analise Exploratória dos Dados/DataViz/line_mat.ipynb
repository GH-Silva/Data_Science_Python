```python
import matplotlib.pyplot as plt
```

### Dados ficticios - Vendas ao longo dos meses

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [150, 200, 180, 300, 250, 400]
```

### Criando o gr√°fico de linhas

```python
plt.figure(figsize=(8, 5))
plt.plot(
    meses,
    vendas,
    marker='o',
    linestyle='-',
    color='blue',
    label='Vendas'
)

### Adicionando r√≥tulos e t√≠tulo ao gr√°fico

plt.xlabel('M√™s')
plt.ylabel('Vendas')
plt.title('Vendas ao longo dos Meses')
plt.legend()
plt.grid(True)
plt.show()
```
