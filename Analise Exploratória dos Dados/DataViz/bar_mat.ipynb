```python
import matplotlib.pyplot as plt
```

### Dados fictícios - Produtos vendidos por vendedor

```python
vendas = ['João', 'Maria', 'Pedro', 'Ana']
qtde_vendida = [45, 60, 30, 55]
```

```python
plt.figure(figsize=(8,5))
plt.bar(
    vendas,
    qtde_vendida,
    color='red'
)

### Adicionando rótulos e título ao gráfico

plt.xlabel('Safra')
plt.ylabel('Vendas')
plt.title('Produtos vendidos por vendedor')
plt.legend()
plt.show()

plt.show()
```
